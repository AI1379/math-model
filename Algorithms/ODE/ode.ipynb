```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
```

```python
# Stiff ODE example
def func1(t, y):
    return -2 * y + 2 * t**2 + 2 * t

# Non-stiff ODE example
def func2(t, y):
    return [
        y[1],
        y[2],
        3 * y[2] + y[1] * y[0]
    ]
```

```python
# Solve func1
from scipy.integrate import solve_ivp
sol = solve_ivp(func1,
                t_span=[0, 0.5],
                y0=[1],
                method='RK45',
                t_eval=np.linspace(0, 0.5, 100),
                )

# Plot the solution of func1
plt.figure()
plt.plot(sol.t, sol.y[0], label='y(t)')
plt.title('Solution of Stiff ODE using RK45')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Non-stiff ODE example
from scipy.integrate import solve_ivp
sol2 = solve_ivp(
    func2,
    t_span=[0, 1],
    y0=[1, 0, -1],
    method="BDF",
    t_eval=np.linspace(0, 1, 100),
)

plt.figure()
plt.plot(sol2.t, sol2.y[0], "-", label="y1(t)", linewidth=2)
plt.plot(sol2.t, sol2.y[1], "--", label="y2(t)", linewidth=2)
plt.plot(sol2.t, sol2.y[2], ":", label="y3(t)", linewidth=2)
plt.title("Solution of Non-stiff ODE using RK45")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
plt.show()
```
